```python
from mass import MassClassifier
from scipy import sparse
import numpy as np
import pandas as pd
```

```python
old=pd.read_csv("0.csv")
```

```python
Y1=y[(~(y==0).all(1))]
```

# setting 1 predicting with ALL features

```python
from catboost import CatBoostClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import MultiLabelBinarizer

ovr = OneVsRestClassifier(estimator=CatBoostClassifier(iterations=600,random_state=0))

import pandas as pd
import numpy as np
import random

ovr.fit(X1,Y1)
from sklearn.neighbors import KNeighborsClassifier
knn = OneVsRestClassifier(estimator=KNeighborsClassifier(n_neighbors=5))

knn.fit(X1, Y1)
```

```python
def score(pred,label):
    plus=np.sum(pred*label,axis=1)
    minus=np.sum(pred*(1-label),axis=1)
    return np.mean((plus-minus)/np.sum(label,axis=1))
```

```python
score(ovr.predict(testdata[:,2,:]),test_label)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=1)
neigh.fit(X1, Y1)
score(neigh.predict(testdata[:,2,:]),test_label)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred_R1 = gnb.fit(X1, Y1[:,0]).predict(testdata[:,2,:])
y_pred_R2 = gnb.fit(X1, Y1[:,1]).predict(testdata[:,2,:])
y_pred_R3 = gnb.fit(X1, Y1[:,2]).predict(testdata[:,2,:])
score(np.concatenate([y_pred_R1.reshape(-1,1),y_pred_R2.reshape(-1,1),y_pred_R3.reshape(-1,1)],axis=1),test_label)
```
